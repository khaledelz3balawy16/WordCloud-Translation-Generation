```
# import and initialize tokenizer and model (text generation)
from transformers import AutoTokenizer, AutoModelWithLMHead
tokenizer = AutoTokenizer.from_pretrained("akhooli/gpt2-small-arabic")
model = AutoModelWithLMHead.from_pretrained("akhooli/gpt2-small-arabic")
```

```
prompt = "القدس مدينة تاريخية تقع في "
```

```
input_ids = tokenizer.encode(prompt, return_tensors='pt').to('cuda')
model.to('cuda');
```

```
%%time
max_length = 504
#set top_k = 40 and num_return_sequences = 3
sample_outputs = model.generate(input_ids, pad_token_id=50256,
                                   do_sample=True,
                                   max_length=max_length,
                                   min_length=max_length,
                                   top_k=40,
                                   num_return_sequences=5)

for i, sample_output in enumerate(sample_outputs):
    print(">> Generated text {}\n\n{}".format(i+1, tokenizer.decode(sample_output.tolist())))
    print('\n---')
```

```
#============= poetry
# initialize tokenizer and model
tokenizer_p = AutoTokenizer.from_pretrained("akhooli/gpt2-small-arabic-poetry")
model_p = AutoModelWithLMHead.from_pretrained("akhooli/gpt2-small-arabic-poetry")
```

```
prompt_p = "على قدر أهل العزم تأتي العزائم - وتأتي على قدر الكرام المكارم."
input_ids_p = tokenizer_p.encode(prompt_p, return_tensors='pt').to('cuda')
model_p.to('cuda');
```

```
%%time
max_length = 120
sample_outputs_p = model_p.generate(input_ids_p, pad_token_id=50256,
                                   do_sample=True,
                                   max_length=max_length,
                                   min_length=max_length,
                                   top_k=40,
                                   num_return_sequences=5)

for i, sample_output in enumerate(sample_outputs_p):
    gen = tokenizer_p.decode(sample_output.tolist())
    print(20*'=', '\n')
    print(f"Sample {i+1}")
    for line in gen.split('.')[:-1]:
        print(line)
    #print(">> Generated poetry {}\n\n{}".format(i+1, tokenizer_p.decode(sample_output.tolist())))
    #print('\n---')
```

```
# from hf blog
# set seed to reproduce results. Feel free to change the seed though to get different results
import torch
torch.manual_seed(42)

# set top_k = 50 and set top_p = 0.95 and num_return_sequences = 3
sample_outputs = model_p.generate(
    input_ids_p,
    do_sample=True,
    max_length=120,
    top_k=50,
    top_p=0.95,
    num_return_sequences=3,
    early_stopping=True,
    min_length=96,
    repetition_penalty=1.25
)

print("Output:\n" + 100 * '-')
for i, sample_output in enumerate(sample_outputs_p):
  print("{}: {}".format(i, tokenizer_p.decode(sample_output, skip_special_tokens=True)))

for i, sample_output in enumerate(sample_outputs_p):
    gen = tokenizer_p.decode(sample_output.tolist())
    print(20*'=', '\n')
    print(f"Sample {i+1}")
    for line in gen.split('.')[:-1]:
        print(line)
```

```
# gpt2-small-arabic-poetry
from simpletransformers.language_generation import LanguageGenerationModel
model = LanguageGenerationModel("gpt2", "akhooli/gpt2-small-arabic-poetry",
                                args={"max_length": 256, "manual_seed": None,
                                      "repetition_penalty":1.3,
                                      "top_p":0.95,"top_k":30, "temperature":0.95
                                     }
                                )
prompts = [
          " .لولا المشقة ساد الناس كلهم - الجود يفقر والإقدام قتال",
     ". ",

]


for prompt in prompts:
    # Generate text using the model. Verbose set to False to prevent logging generated sequences.
    generated = model.generate(prompt, verbose=False)
    #print('raw:', generated)
    print(20*'=', '\n')
    for line in generated[0].split('.')[:-1]:
        print(line)
```

```
#====================================== Multiligual classification =======================
# this is an XLM-Roberta (100 languages) model fine-tuned on Arabic 3-class sentiment analysis dataset
# dataset: https://www.kaggle.com/abedkhooli/arabic-100k-reviews
# see https://arxiv.org/abs/1911.02116 (https://arxiv.org/pdf/1911.02116.pdf Appendix A)  for supported languages
# disclaimer: model not fully trained, mixed sentiment may not be accurately predicted (training data is based on rating of 3 ouot of 5)
```

```
# if not installed, install transformers (uncomment next line)
#!pip install -U -q transformers

#initiate pipeline for sentiment analysis with specific model
from transformers import pipeline
# this will download 2 GB
nlp = pipeline("sentiment-analysis", model='akhooli/xlm-r-large-arabic-sent')
```

```
# 0 mixed, 1 negative, 2 positive
print(nlp("I hate you")) # En negative
print(nlp("I love you")) # En positive
print(nlp("المنتج جيد جدا ولكنه وصل بحالة غير ممتازة")) # Ar mixed
# Ar positive
print(nlp("موقع المكان جميل جداً البناء قديم جداً ، يقدم الاكل طازج ولذيذ، معامله وخدمه ممتازه، انصح جداً بزياره المكان ، انا زبون دائم، وايضاً قريب على كنيسه المهد"))
# Ar negative
print(nlp("ببساطه اسوأ طيران في العالم , الطياره قديمه جدا جدا والطاقم والموظفين عند الوزن سيئين جدااااااااا الى ابعد الحدود + فقدو شنته السفر + تأخير طياره نصيحه من مجرب لا تحجز على الطيران الملكيه اسوأ طيران قد ركبته"))
# Ar and En positive
print(nlp("كانت الخدمة super!"))
```

```
print(nlp("الخدمة كانتsuper"))
print(nlp("الخدمة في المقهى لم تكن bad بل كانت جيدة"))
print(nlp("إجمالا، الخدمة لم تكن سيئة"))
print(nlp("على أية حال،الخدمة ما كانت جيدة"))
```

```
#====================================== Toxic/hate speech classification =======================
# this is an XLM-Roberta (100 languages) model fine-tuned on a subset of Jigsaw dataset (machine translated into Arabic)
# Jigsaw dataset: https://www.kaggle.com/c/jigsaw-multilingual-toxic-comment-classification/data
# May work for other languages (zero shot)
# see https://arxiv.org/abs/1911.02116 (https://arxiv.org/pdf/1911.02116.pdf Appendix A)  for supported languages
# disclaimer: model not fully trained, train data is limited and no original text (machine translation)
```

```
# if not installed, install transformers (uncomment next line)
!pip install -U -q transformers

#initiate pipeline for sentiment analysis with specific model
from transformers import pipeline
# this will download 2 GB
nlp = pipeline("sentiment-analysis", model='akhooli/xlm-r-large-arabic-toxic')
```

```
# 0 non-toxic, 1 toxic (hate)
print(nlp("أنت غير مؤدب. إلى الجحيم!"))
print(nlp("أختلف معك في هذا الرأي، فالبشر ليسوا من نسل القرود ودارون ليس على حق."))
```
